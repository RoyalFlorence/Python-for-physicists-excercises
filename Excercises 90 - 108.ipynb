```python
import numpy as np
import matplotlib.pyplot as plt

#Ex. 90/91
#Forward difference method
#Calculates the derivate at point x + dx/2

#Define a function to take the derivative of
def f(ii):
    return np.sin(ii)
#Define an array of x and f(x) values
#x-values
x = np.linspace(0,4*np.pi,100)
#f(x)-values
fx = [f(x[ii]) for ii in range(0, len(x))]


#Define a function which contains the forward difference method
#Takes the values of x and f(x) as input and f'(x) as output.

def ForwardMethod(xvalues, fxvalues):
    diff = np.zeros(len(xvalues))
    delta_x = xvalues[1] - xvalues[0]
    for ii in range(0, len(xvalues)-1):
        diff[ii] = (fxvalues[ii+1] - fxvalues[ii])/delta_x
    diff[len(xvalues)-1] = diff[len(xvalues)-2]
    return diff

plt.plot(x, fx, label='function value')
plt.plot(x,ForwardMethod(x, fx), label='forward derivative')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Ex. 91/92
#Forward difference method
#Compare the method with various step sizes
n1 = 5
n2 = 10
n3 = 20
x_values1 = np.linspace(0, 5, n1)
x_values2 = np.linspace(0, 5, n2)
x_values3 = np.linspace(0, 5, n3)
print('length x_values3 is', len(x_values3))
def f(x):
    return np.exp(x)
fun_values1 = [f(x_values1[ii]) for ii in range(0, len(x_values1))]
fun_values2 = [f(x_values2[ii]) for ii in range(0, len(x_values2))]
fun_values3 = [f(x_values3[ii]) for ii in range(0, len(x_values3))]
print('length x_values3 is', len(x_values3), 'and length fun_values3 is', len(fun_values3))
def forward_derivative(x, fun):
    stepsize = x[1] - x[0]
    diff = np.zeros(len(x))
    for ii in range(0, len(x)-1):
        diff[ii] = (fun[ii+1] - fun[ii])/stepsize
    diff[len(x)-1] = diff[len(x)-2]
    return diff

#Plot all three difference stepsizes
fig, ax = plt.subplots(nrows=1, ncols=3, constrained_layout=True)
ax[0].plot(x_values1, forward_derivative(x_values1, fun_values1))
ax[0].set_title('n1')
ax[1].plot(x_values2, forward_derivative(x_values2, fun_values2))
ax[1].set_title('n2')
ax[2].plot(x_values3, forward_derivative(x_values3, fun_values3))
ax[2].set_title('n3')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Ex. 93
#Midpoint rule differentiation

#Define a function to take the derivative of
def f(ii):
    return np.exp(ii)
#Define an array of x and f(x) values
#x-values
x_values = np.linspace(0,2,100)
#f(x)-values
fun_values = [f(x_values[ii]) for ii in range(0, len(x_values))]

def symmetric_derivative(x, fun):
    
    diff = np.zeros(len(x))
    for ii in range(1, len(x)-1):
        stepsize = x[ii+1] - x[ii]
        diff[ii] = (fun[ii+1] - fun[ii-1])/(2*stepsize)
    diff[0] = diff[1]
    diff[-1] = diff[-2]
    return diff
#print(symmetric_derivative(x_values,fun_values))
plt.plot(x_values, fun_values, 'r', label='f(x)')
plt.plot(x_values, symmetric_derivative(x_values, fun_values), 'b--', label='f,(x)')
plt.suptitle('Midpoint Difference Method(waarom is de afgeleide zo afgeschaald)')
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Ex. 94
#Multiple order midpoint differentiation
x_val = np.arange(0,np.pi*3, 0.01)
def f(x):
    return np.cos(x)
y_val = [f(x_val[ii]) for ii in range(0, len(x_val))]

order = 4

def symmetric_derivative(x, y):
    stepsize = x[1]-x[0]
    n_diff = np.zeros(len(y))
    for ii in range(1, len(x)-1):
            n_diff[ii] = (y[ii+1] - y[ii-1]) /  (2*stepsize)
    return n_diff

def nth_derivative(x, y, n):
    n_diff = y
    nn = 0
    while nn < n:
        n_diff = symmetric_derivative(x, n_diff)
        nn += 1
    for ii in range(0, n+1):
        n_diff[ii] = n_diff[ii+n+1]
    for ii in range(len(y)-n, len(y)):
        n_diff[ii] = n_diff[ii-1]
    return n_diff

plt.plot(x_val,y_val, 'b--', label='function')
n_diff_val = nth_derivative(x_val, y_val, order)
plt.plot(x_val, n_diff_val, 'r', label='nth derivative')
plt.legend(bbox_to_anchor=(1.35,1.01))
plt.grid()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

#Ex. 97

#midpoint integration
def midpoint_integration(fun, a, b, N):
    pre_value = (b-a)/N
    x_mid = [(ii+0.5)*((b-a)/N) for ii in range(0, N+1)]
    fun_mid = [fun(x_mid[ii]) for ii in range(0, len(x_mid))]
    fun_mid_sum = sum(fun_mid)
    return fun_mid_sum * pre_value

#integral boundaries
a = -2
b = 2
N = 10

#function to integrate over
def function(x):
    return x**2 - 2

#quad integral value
int_quad = integrate.quad(function, a, b)[0]

#list for midpoint integral values for each n in range N
midpoint_integral_N_list = np.zeros(N)
difference_midpoint_quad = np.zeros(N)
N_list = np.zeros(N)
for nn in range(1, N):
    N_list[nn] = nn
    midpoint_integral_N_list[nn] = midpoint_integration(function, a, b, nn)
    difference_midpoint_quad[nn] = midpoint_integral_N_list[nn] - int_quad
plt.plot(N_list, difference_midpoint_quad)


    
```
